# 1 - Stochastic Processes

Stochastic processes can be defined as a collection of random variables indexed or ordered in time. 

The data observed in time is a particular realization of infinite possibilities. The sample is to the population what a realization is to the adjacent stochastic process.

___

## 1.1 - Stationary Stochastic Processes

In a stationary stochastic process, the mean, variance, and covariance are the same no matter where they are measured. The mean and variance are constant, while the covariance between two points depends only on the lag between these periods.

Such series tend to return to the mean according to the autocovariance. Low autocovariance will make the return fast, while high autocovariance will make the return slow.

A pure stochastic or white noise process has zero mean, constant variance, and is serially uncorrelated. Furthermore, the probability distributions are constant over time.

Finally, stationarity is an aspect of the adjacent stochastic process, not the available realization.

___

## 1.2 - Non-Stationary Stochastic Processes

A stochastic process is not stationary when the mean or the variance or both are not constant in time.

Random walk can be understood as yesterday's value plus a random shock.

### Random Walk Without Displacement

A random walk without displacement is a non-stationary stochastic process. 

## $  Y_{t} = Y_{t-1} + \mu_{t} $ 
* non-stationary 

## $  \Delta_{t} = Y_{t} - Y_{t-1} = \mu_{t}  $
* stationary

## $  Y_{t} = Y_{0} + \sum\mu_{t} $
* $\sum\mu_{t} $ Stochastic trend with infinite memory. 

##  $VAR(Y_{t}) = t\sigma^2 $
* non-stationary

The first difference of a random walk without displacement is stationary, however, the variance increases indefinitely with increasing time.

### Random Walk with Displacement

A random walk with displacement is also a non-stationary stochastic process. 

## $  Y_{t} = \delta + Y_{t-1} + \mu_{t} $ 
* non-stationary

## $  \Delta_{t} = Y_{t} - Y_{t-1} = \delta + \mu_{t}  $
* non-stationary

## $  Y_{t} = \delta + Y_{0} + \sum\mu_{t} $
* $\delta $ displacement parameter 


##  $VAR(Y_{t}) = t\sigma^2 $
* non-stationary

Both the first difference and the variance are not constant.

___

## 1.3 - Unit Root Stochastic Processes

## $  Y_{t} = \rho Y_{t-1} + \mu_{t} $ 

## $  \Delta_{t} = \rho Y_{t} - Y_{t-1} + \mu_{t} $
### $ \Delta_{t} = Y_{t}(1 - p) + \mu_{t} $
### $ \Delta_{t} = \alpha Y_{t} + \mu_{t} $

 * $ \rho = 1 $
 * $ \alpha = 0 $
 * $ \Delta_{t} = \mu_{t} $
 * $ non-stationary $
 
 
 * $ \rho = 0 $
 * $ \alpha = 1 $
 * $ \Delta_{t} = Y_{t} + \mu_{t} $
 * $ non-stationary $
 
 
 * $ \rho > 0 $
 * $ \alpha < 1 $
 * $ \Delta_{t} = \alpha Y_{t} + \mu_{t} $
 * $ stationary $
 

___

## 1.3 - Integrated Stochastic Processes

The random walk is a specific case within stochastic processes.

Integration or Order of Integration is the number of times a series needs to be differentiated to become stationary.

## Unit Root Test

### Dickey-Fuller Test

## $  Y_{t} = \rho Y_{t-1} + \mu_{t} $ 

## $  \Delta_{t} = \rho Y_{t} - Y_{t-1} + \mu_{t} $
## $  \Delta_{t} = (\rho - 1) Y_{t-1} + \mu_{t} $

## $  \Delta_{t} = \delta Y_{t-1} + \mu_{t} $

$ \rho $ is between -1 and 1.

$ \rho = 1 $ implies $ \delta = 0 $

**The test consists of testing $ \rho = 0 $ as the null hypothesis**

$ {H}_{0} => \rho = 0 $

$ {H}_{1} => \rho < 0 $




Test:

1) Take the first difference.

2) Regress $ Y $ against $ {Y}_{t-1} $.

3) Validate null hypothesis
* If $ \rho = 0$ the series is not stationary
* If $ \rho < 0$ the series is stationary

The DF test assumes that the errors are uncorrelated.

### Augmented Dickey-Fuller Test

The augmented DF test assumes that the errors are correlated.

The difference between the DF test and the ADF test lies in the addition of lagged values ​​of the variable Y

Using the equation  $ Y_{t} = {\beta}_{0} + {\beta}_{1}t + {\delta}{Y}_{t-1} + {\mu}_{t}$ as an example, we have:

### $ \Delta Y_{t} = {\beta}_{0} + {\beta}_{1}t + {\delta}{Y}_{t-1} + {\alpha}_{1} {\Delta Y}_{t-1} + 
{\alpha}_{2} {\Delta Y}_{t-2} + ... + {\alpha}_{i} {\Delta Y}_{t-i} + {\epsilon}_{i}$

### $ \Delta Y_{t} = {\beta}_{0} + {\beta}_{1}t + {\delta}{Y}_{t-1} + \sum{\alpha}_{i}\Delta Y_{t-i} + {\epsilon}_{i} $


The equation representing the type of random walk could be:

### $ Y_{t} = {\beta}_{1}{Y}_{t-1} + {\mu}_{t}$ 


### $ Y_{t} = {\beta}_{0}  + {\beta}_{1}{Y}_{t-1} + {\mu}_{t}$

## Cointegration Test

Two variables are cointegrated when there is a long-term or equilibrium relationship between them. Economic theory is often expressed in terms of equilibrium, such as the balance between supply and demand or the quantity theory of Fisher's money.

In practical terms, cointegration occurs when we regress a time series with a unit root against another time series also with a unit root. That is, two series I(1) (non-stationary) combined linearly are I(0) stationary. The relationship between income and consumption is another example.

### $ poupança I(0) = renda I(1)  - consumo I(1) $ 

A cointegration test can be a pretest to avoid spurious regressions.

### Cointegration Test

The cointegration test can be performed by applying the Dickey-Fuller or DFA test on the residuals of the equation. These tests are known as Engle-Granger and Augmented Engle-Granger.

# 3 - ARMA

The ARMA process can be seen as an ARIMA process with degree of integration $ I = 0 $, that is, it is an already stationary series.

#### ARMA (p,q)

    * p = number of autoregressive terms
    * q = number of moving average terms

AR autoregressive process

### AR(1) => $ (Y_t - \tilde{Y_t}) = \alpha_1 (T_{t-1} - \tilde{Y_t}) + \mu_t   $

### AR(2) => $ (Y_t - \tilde{Y_t}) = \alpha_1 (T_{t-1} - \tilde{Y_t}) + \alpha_2 (T_{t-2} - \tilde{Y_t}) + \mu_t   $

MA moving average process

### MA(1) => $ Y_t = \mu + \beta_0\mu_t + \beta_1\mu_{t-1}  $

### MA(2) => $ Y_t = \mu + \beta_0\mu_t + \beta_1\mu_{t-1} + \beta_2\mu_{t-2}  $

Autoregressive moving averages process

### ARMA(1,1) => $ Y_t = \sigma + \alpha Y_{t-1}+ \beta_0\mu_t + \beta_1\mu_{t-1}  $

___

# 4 - ARIMA

The ARIMA process can be understood, in a simplified way, as a generalization of the ARMA process. An ARIMA(1,0,1) process is equivalent to an ARMA(1,1) process.

#### ARIMA (p, d,  q)

    * p = number of autoregressive terms
    * d = number of times the series is differentiated until it becomes stationary
    * q = number of moving average terms

Processo auto-regressivo AR

### AR(1) => $ (Y_t - \tilde{Y_t}) = \alpha_1 (T_{t-1} - \tilde{Y_t}) + \mu_t   $

### AR(2) => $ (Y_t - \tilde{Y_t}) = \alpha_1 (T_{t-1} - \tilde{Y_t}) + \alpha_2 (T_{t-2} - \tilde{Y_t}) + \mu_t   $

Proceso de integração

$ \Delta Y_t  = Y_t - Y_{t-1} $

Processo de média móvel MA

### MA(1) => $ Y_t = \mu + \beta_0\mu_t + \beta_1\mu_{t-1}  $

### MA(2) => $ Y_t = \mu + \beta_0\mu_t + \beta_1\mu_{t-1} + \beta_2\mu_{t-2}  $

Processo auto-regressivo de médias móveis

### ARIMA(1,1,1) => $ \Delta Y_t = \sigma + \alpha \Delta Y_{t-1}+ \beta_0\mu_t + \beta_1\mu_{t-1}  $

___

## Box-Jenkins

When dealing with a time series, such as GDP for example, how can we know if it follows an AR process or a pure MA process or an ARMA or ARIMA process? The Box-Jenkins method helps to answer this question using four steps to identify and estimate the model.

1) Identification
    
    * ACF sample correlation function
    * PACF partial sample correlation function
        * Correlograms
            * ACF and PACF against lag extension
        
Partial autocorrelation is analogous to the partial correlation coefficient. In other words, it is the relationship between $Y_t$ and $Y_{t-\alpha}$ after removing the effect of the intermediate $Y_i$.


2) Estimation of parameters

After identifying the p, d, q values ​​of the model, the next step is to estimate the parameters. Sometimes the estimation can be by the simple least squares method, however, it may be necessary to use nonlinear estimation methods. In practice, the calculation step is performed by statistical packages.


3) Diagnosis

Are the estimated noises white noise? If not, the estimation process must be restarted, making the method iterative.


4) Forecast

The main object of econometric models is to make predictions and the ARIMA model usually obtains very reliable predictions in the short term when compared to traditional econometric methods.

# 5 - SARIMA

#### SARIMA (p, d,  q)(P, D, Q)


SARIMA models are models of the univariate time series model class. The acronym SARIMA stands for Integrated AutoRegressive Models of Moving Averages with Seasonality. They are very useful models to generate time series forecasts when, in general, predictor variables are not available. The most interesting aspect of this type of approach is precisely to place the seasonal part of the series within the model.

SARIMA MODELS ESTIMATION PROCESS

* Visualize the data to identify non-standard observations (outliers or missing data), examine whether there is a trend and/or seasonality through specific graphs
* If necessary, transform the data to stabilize the variance (data logarithm, variance or return, for example)
* Test whether the data is stationary. If it has a unit root, it is necessary to differentiate the data until they become stationary. For this, it is tested again whether the differentiated series has become stationary.
If there is only seasonality, do seasonal differentiation (e.g. for monthly data do 12th order differentiation)
If there is only trend, make only the first difference.
If there is trend and seasonality, apply the first difference in the data to treat the trend and after that the seasonal differentiation (the case of Coca-Cola's quarterly earnings)
If there is no trend or seasonality, do not make any differentiation.
* Examine the partial autocorrelation (FAC) and partial autocorrelation (FACP) functions to determine the maximum orders P and Q for the AR and MA components of both the regular and seasonal parts of the stationary series:
non-seasonal terms: examine the first lags (1,2,3,..) of the FAC and FACP. We maintain the same pattern already studied, that is, the FAC defines MA terms and the FACP defines AR terms
seasonal terms: examine patterns in lags that are multiples of the series periodicity. For example, for monthly data, check lags 12, 24, 36 (you will probably need to check the first two or three multiple lags of the periodicity). Evaluate FAC and FACP for seasonal lags the same way you did for non-seasonal lags
* Estimate all combinations for p, d and q in the regular part and P, D and Q in the seasonal part. Here, both d and D will be fixed and equal to the number of times needed to make the original series stationary in the regular part and in the seasonal part, respectively.
* Choose from all the models estimated in the previous step, the model with the lowest AIC and/or BIC.
* Examine whether the residuals behave as white noise. Otherwise, return to step 3 or 4.
Test autocorrelation in residuals
Test for conditional heteroscedasticity
Check the probability distribution
* Since the residuals are white noise, get the predictions.

# <font size="5">2 - Stationarity: Rolling Statistics, Autocorrelação e Teste Dickey-Fuller</font>

### Verificando a Estacionariedade

Podemos verificar a estacionariedade de uma série temporal da seguinte forma:

- **Plot das Rolling Statistics (Estatísticas Móveis)**: Podemos criar um plot da média móvel e variância móvel (ou desvio padrão, que é a raiz quadrada da variância) e ver se variam com o tempo. A média / variância móvel é, para qualquer instante 't', a média / variância do último ano, ou seja, os últimos 12 meses. Curiosidade: mostramos como aplicar Rolling Statistics a dados em tempo real usando o Apache Spark no curso Machine Learning e IA em Ambientes Distribuídos da Formação Engenheiro de Dados.


- **Autocorrelação com Gráficos ACF e PACF**: Se a série temporal for estacionária, os gráficos ACF / PACF (gráficos de autocorrelação da série) mostrarão um padrão específico o qual discutiremos mais adiante. Curiosidade: mostramos como calcular a correlação em R no curso Análise Estatística Para Data Science II da Formação Análise Estatística.


- **Teste Dickey-Fuller**: Este é um dos testes estatísticos para verificar a estacionariedade. Aqui, a hipótese nula é que a série temporal não é estacionária. Os resultados do teste incluem uma estatística de teste e alguns valores críticos para os níveis de confiança das diferenças. Se a 'Estatística de teste' for menor que o 'valor crítico', podemos rejeitar a hipótese nula e dizer que a série é estacionária. Curiosidade: mostramos como realizar o Teste Dickey-Fuller em R no curso Análise Estatística Para Data Science II da Formação Análise Estatística.

Usaremos os 3 métodos para checar se a nossa série é estacionária ou não.

## Rolling Statistcs

```python
# Prepara os dados
x = dataset['mes'].values
y = dataset['Serie_Original_Passageiros'].values

# Determinando estatísticas móveis
rolmean = dataset.rolling(window = 12).mean()
rolstd = dataset.rolling(window = 12).std()

# Plot das estatísticas móveis
fig, ax = plt.subplots(figsize = (15,7))
orig = plt.plot(y, color = 'blue', label = 'Original')
mean = plt.plot(rolmean, color = 'red', label = 'Média Móvel')
std = plt.plot(rolstd, color = 'black', label = 'Desvio Padrão')
plt.legend(loc = 'best')
plt.title('Estatísticas Móveis - Média e Desvio Padrão')
plt.show()
```

### ACF e PACF

```python
# Plots
plt.rcParams.update({'figure.figsize': (16,10)})

# Plot do gráfico ACF
# https://www.statsmodels.org/stable/generated/statsmodels.graphics.tsaplots.plot_acf.html
plt.subplot(211)
plot_acf(dataSeries.Serie_Original_Passageiros, ax = plt.gca(), lags = 30)

# Plot do gráfico PACF
# https://www.statsmodels.org/stable/generated/statsmodels.graphics.tsaplots.plot_pacf.html
plt.subplot(212)
plot_pacf(dataSeries.Serie_Original_Passageiros, ax = plt.gca(), lags = 30)
plt.show()
```

### Teste Dickey-Fuller aumentado

- A intuição por trás do teste é que, se a série for integrada, o nível de atraso da série y (t-1) não fornecerá informações relevantes na previsão da mudança em y (t).


- Hipótese nula: a série temporal não é estacionária.


- Rejeitar a hipótese nula (ou seja, um valor-p abaixo de 0.05) indicará estacionaridade.

```python
# Teste Dickey-Fuller

# Print
print('\nResultado do Teste Dickey-Fuller:\n')

# Teste
dfteste = adfuller(dataSeries.Serie_Original_Passageiros, autolag = 'AIC')

# Formatando a saída
dfsaida = pd.Series(dfteste[0:4], index = ['Estatística do Teste',
                                           'Valor-p',
                                           'Número de Lags Consideradas',
                                           'Número de Observações Usadas'])

# Loop por cada item da saída do teste
for key,value in dfteste[4].items():
    dfsaida['Valor Crítico (%s)'%key] = value

# Print
print (dfsaida)
```

## Function to Test Stationarity

```python
def stationarity_test(serie):
    
    # Calcula estatísticas móveis
    rolmean = serie.rolling(window = 12).mean()
    rolstd = serie.rolling(window = 12).std()

    # Plot das estatísticas móveis
    fig, ax = plt.subplots(figsize = (15,7))
    orig = plt.plot(serie, color = 'blue', label = 'Original')
    mean = plt.plot(rolmean, color = 'red', label = 'Média Móvel')
    std = plt.plot(rolstd, color = 'black', label = 'Desvio Padrão')
    plt.legend(loc = 'best')
    plt.title('Estatísticas Móveis - Média e Desvio Padrão')
    plt.show()
    
    # Teste Dickey-Fuller:
    # Print
    print('\nResultado do Teste Dickey-Fuller:\n')

    # Teste
    dfteste = adfuller(dataSeries.Serie_Original_Passageiros, autolag = 'AIC')

    # Formatando a saída
    dfsaida = pd.Series(dfteste[0:4], index = ['Estatística do Teste',
                                               'Valor-p',
                                               'Número de Lags Consideradas',
                                               'Número de Observações Usadas'])

    # Loop por cada item da saída do teste
    for key, value in dfteste[4].items():
        dfsaida['Valor Crítico (%s)'%key] = value

    # Print
    print (dfsaida)
    
    # Testa o valor-p
    print ('\nConclusão:')
    if dfsaida[1] > 0.05:
        print('\nO valor-p é maior que 0.05 e, portanto, não temos evidências para rejeitar a hipótese nula.')
        print('Essa série provavelmente não é estacionária.')
    else:
        print('\nO valor-p é menor que 0.05 e, portanto, temos evidências para rejeitar a hipótese nula.')
        print('Essa série provavelmente é estacionária.')
```

```python
# Check
stationarity_test(dataSeries)
```

# <font size="5">4 - Smoothing </font>

A suavização (Smoothing) é basicamente uma técnica usada para ver a tendência de longo prazo nos dados, diminuindo os efeitos dos componentes periódicos / sazonais dos dados. Basicamente, usamos suavização quando queremos remover as flutuações nos dados e focamos apenas em preservar as tendências de longo prazo.

```python
# Vamos criar uma cópia da série original pois isso será importante mais tarde
df_serie = dataSeries.copy()

# Agora definimos uma janela de 12 meses da série temporal para calcular a média móvel e guardamos no mesmo dataset
dataSeries['Serie_Suavizada_Passageiros'] = dataSeries.rolling(window = 12).mean()

# drop na
dataSeries.dropna(inplace = True)

# View
dataSeries.head()
```

## Function to Test Stationarity

```python
# Função para testar a estacionaridade
def stationarity_testII(serie):
    
    # Calcula estatísticas móveis
    rolmean = serie.rolling(window = 12).mean()
    rolstd = serie.rolling(window = 12).std()

    # Plot das estatísticas móveis
    fig, ax = plt.subplots(figsize = (15,7))
    orig = plt.plot(serie, color = 'blue', label = 'Original')
    mean = plt.plot(rolmean, color = 'red', label = 'Média Móvel')
    std = plt.plot(rolstd, color = 'black', label = 'Desvio Padrão')
    plt.legend(loc = 'best')
    plt.title('Estatísticas Móveis - Média e Desvio Padrão')
    plt.show()
    
    # Teste Dickey-Fuller:
    # Print
    print('\nResultado do Teste Dickey-Fuller:\n')

    # Teste
    dfteste = adfuller(serie, autolag = 'AIC')

    # Formatando a saída
    dfsaida = pd.Series(dfteste[0:4], index = ['Estatística do Teste',
                                               'Valor-p',
                                               'Número de Lags Consideradas',
                                               'Número de Observações Usadas'])

    # Loop por cada item da saída do teste
    for key, value in dfteste[4].items():
        dfsaida['Valor Crítico (%s)'%key] = value

    # Print
    print (dfsaida)
    
    # Testa o valor-p
    print ('\nConclusão:')
    if dfsaida[1] > 0.05:
        print('\nO valor-p é maior que 0.05 e, portanto, não temos evidências para rejeitar a hipótese nula.')
        print('Essa série provavelmente não é estacionária.')
    else:
        print('\nO valor-p é menor que 0.05 e, portanto, temos evidências para rejeitar a hipótese nula.')
        print('Essa série provavelmente é estacionária.')
```

```python
stationarity_testII(diff_original_suavizada)
```

# <font size="5">5 - Differentiation </font>

Nesta técnica, tomamos a diferença da observação em um determinado instante com a do instante anterior. A diferenciação é realizada subtraindo a observação anterior da observação atual:

difference(t) = observation(t) - observation(t-1)

A inversão do processo é necessária quando uma previsão deve ser convertida novamente na escala original. Este processo pode ser revertido adicionando a observação no passo anterior ao valor da diferença.

inverted(t) = differenced(t) + observation(t-1)

**Ordem da Diferença**

Alguma estrutura temporal ainda pode existir após a execução de uma operação de diferenciação, como no caso de uma tendência não linear.

Como tal, o processo de diferenciação pode ser repetido mais de uma vez até que toda a dependência temporal seja removida.

O número de vezes que a diferenciação é realizada é chamado de ordem da diferença. Na primeira diferenciação temos a Diferenciação de Primeira Ordem e assim por diante.

### Diferenciação Para Remover Tendência

Uma tendência torna uma série temporal não estacionária. Isso tem o efeito de variar o valor médio da série temporal ao longo do tempo. Veja 2 exemplos de como aplicar a diferenciação para remover tendência.

```python
# Função para diferenciação, se desejar:
def diffFunc(dataset, interval = 1):
    diff = list()
    for i in range(interval, len(dataset)):
        value = dataset[i] - dataset[i - interval]
        diff.append(value)
    return diff
```

```python
# Diferenciação Para Remover Tendência usando a série com transformação de log
dados_serie_diff1 = diffFunc(np.log(dataSeries.Serie_Original_Passageiros), 1)

# Converte em dataframe, a lista gerada ao aplicar a diferenciação
dados_serie_diff1_df = pd.DataFrame(dados_serie_diff1) 

fig, ax = plt.subplots(figsize = (15,7))
plt.plot(dados_serie_diff1, color = 'blue')
plt.title('Differentiation to Remove Trend')
plt.show()
```

```python
# Vamos testar a estacionaridade da série após a diferenciação para remover a tendência
stationarity_testII(dados_serie_diff1_df)
```

### Diferenciação Para Remover Sazonalidade

Variação sazonal, ou sazonalidade, são ciclos que se repetem regularmente ao longo do tempo.

Um padrão de repetição dentro de cada ano é conhecido como variação sazonal, embora o termo seja aplicado de maneira mais geral a padrões repetidos em qualquer período fixo.

Existem muitos tipos de sazonalidade. Alguns exemplos óbvios incluem: diariamente, semanalmente, mensalmente, anualmente e assim por diante. Sendo assim, identificar se há um componente de sazonalidade na série temporal é subjetivo.

A abordagem mais simples para determinar se há um aspecto de sazonalidade é plotar e revisar seus dados, talvez em escalas diferentes e com a adição de linhas de tendência.

O exemplo abaixo aplica a função diffFunc() à última versão da nossa série até aqui, com transformacão de log e diferenciação para remoer tendência. 

```python
# Diferenciação para remover sazonalidade
# Observe o valor 2. Representa o intervalo para calcular a diferenciação. 
# Colocamos o valor 2 apenas para mostrar que é possível, mas se usarmos o valor 1 como usamos na 
# primeira diferenciação, o resultado é tão bom quanto.
dados_serie_diff2 = diffFunc(dados_serie_diff1, 2)

# Converte aem dataframe, a lista gerada ao aplicar a diferenciação
dados_serie_final = pd.DataFrame(dados_serie_diff2) 

fig, ax = plt.subplots(figsize = (15,7))
plt.plot(dados_serie_diff2, color = 'blue')
plt.title('Differentiation to Remove Season')
plt.show()
```

# <font size="5">2 - Exponential Smoothing </font>

**- Exponential Smoothing**

A "suavização exponencial" foi proposta no final da década de 1950 (Brown, 1959; Holt, 1957; Winters, 1960) e motivou alguns dos métodos de previsão mais bem-sucedidos. As previsões produzidas usando métodos de suavização exponencial são médias ponderadas de observações passadas, com os pesos decaindo exponencialmente à medida que as observações "envelhecem". Em outras palavras, quanto mais recente a observação, maior o peso associado. Essa estrutura gera previsões confiáveis rapidamente e para uma ampla gama de séries temporais, o que é uma grande vantagem e de grande importância para muitas aplicações comerciais.

```python
# Função para testar a estacionaridade
def testa_estacionaridade(serie):
    
    # Calcula estatísticas móveis
    rolmean = serie.rolling(window = 12).mean()
    rolstd = serie.rolling(window = 12).std()

    # Plot das estatísticas móveis
    fig, ax = plt.subplots(figsize = (15,7))

    orig = plt.plot(serie, color = 'blue', label = 'Original')
    mean = plt.plot(rolmean, color = 'red', label = 'Média Móvel')
    std = plt.plot(rolstd, color = 'black', label = 'Desvio Padrão')
    plt.legend(loc = 'best')
    plt.title('Estatísticas Móveis - Média e Desvio Padrão')
    plt.show()
    
    # Teste Dickey-Fuller:
    # Print
    print('\nResultado do Teste Dickey-Fuller:\n')

    # Teste
    dfteste = adfuller(serie, autolag = 'AIC')

    # Formatando a saída
    dfsaida = pd.Series(dfteste[0:4], index = ['Estatística do Teste',
                                               'Valor-p',
                                               'Número de Lags Consideradas',
                                               'Número de Observações Usadas'])

    # Loop por cada item da saída do teste
    for key, value in dfteste[4].items():
        dfsaida['Valor Crítico (%s)'%key] = value

    # Print
    print (dfsaida)
    
    # Testa o valor-p
    print ('\nConclusão:')
    if dfsaida[1] > 0.05:
        print('\nO valor-p é maior que 0.05 e, portanto, não temos evidências para rejeitar a hipótese nula.')
        print('Essa série provavelmente não é estacionária.')
    else:
        print('\nO valor-p é menor que 0.05 e, portanto, temos evidências para rejeitar a hipótese nula.')
        print('Essa série provavelmente é estacionária.')
```

```python
# Aplica a função 
testa_estacionaridade(dados_treino['Count'])
```

Em Estatística e Econometria, e em particular na análise de séries temporais, um modelo de média móvel integrada auto-regressiva (ARIMA) é uma generalização de um modelo de média móvel auto-regressiva (ARMA). 

Ambos os modelos são ajustados a dados de séries temporais para melhor entender os dados ou para prever pontos futuros na série (previsão). Os modelos ARIMA são aplicados em alguns casos em que os dados mostram evidências de não estacionariedade, onde uma etapa inicial de diferenciação (correspondente à parte "integrada" do modelo) pode ser aplicada uma ou mais vezes para eliminar a não estacionariedade. 

A parte AR do ARIMA indica que a variável de interesse em evolução é regredida com seus próprios valores defasados (isto é, anteriores). A parte MA indica que o erro de regressão é na verdade uma combinação linear de termos de erro cujos valores ocorreram contemporaneamente e em vários momentos no passado. O I (para "integrado") indica que os valores dos dados foram substituídos pela diferença entre seus valores e os valores anteriores (e esse processo de diferenciação pode ter sido executado mais de uma vez). O objetivo de cada um desses recursos é fazer com que o modelo ajuste os dados da melhor maneira possível.

Modelos ARIMA não sazonais são geralmente designados ARIMA(p, d, q), em que os parâmetros p, d e q são números inteiros não negativos, p é a ordem (número de intervalos de tempo) do modelo autoregressivo, d é o grau de diferenciação (o número de vezes que os dados tiveram valores passados subtraídos) e q é a ordem do modelo de média móvel. 


```python
# Função para testar a estacionaridade
def testa_estacionaridade(serie):
    
    # Calcula estatísticas móveis
    rolmean = serie.rolling(window = 12).mean()
    rolstd = serie.rolling(window = 12).std()

    # Plot das estatísticas móveis
    fig, ax = plt.subplots(figsize = (15,7))

    orig = plt.plot(serie, color = 'blue', label = 'Original')
    mean = plt.plot(rolmean, color = 'red', label = 'Média Móvel')
    std = plt.plot(rolstd, color = 'black', label = 'Desvio Padrão')
    plt.legend(loc = 'best')
    plt.title('Estatísticas Móveis - Média e Desvio Padrão')
    plt.show()
    
    # Teste Dickey-Fuller:
    # Print
    print('\nResultado do Teste Dickey-Fuller:\n')

    # Teste
    dfteste = adfuller(serie, autolag = 'AIC')

    # Formatando a saída
    dfsaida = pd.Series(dfteste[0:4], index = ['Estatística do Teste',
                                               'Valor-p',
                                               'Número de Lags Consideradas',
                                               'Número de Observações Usadas'])

    # Loop por cada item da saída do teste
    for key, value in dfteste[4].items():
        dfsaida['Valor Crítico (%s)'%key] = value

    # Print
    print (dfsaida)
    
    # Testa o valor-p
    print ('\nConclusão:')
    if dfsaida[1] > 0.05:
        print('\nO valor-p é maior que 0.05 e, portanto, não temos evidências para rejeitar a hipótese nula.')
        print('Essa série provavelmente não é estacionária.')
    else:
        print('\nO valor-p é menor que 0.05 e, portanto, temos evidências para rejeitar a hipótese nula.')
        print('Essa série provavelmente é estacionária.')
```

```python
# Aplica a função 
testa_estacionaridade(dados_treino['Count'])
```

## Determinando o Valor do Parâmetro d para o Modelo ARIMA

O objetivo da diferenciação é tornar a série temporal estacionária.

Uma série super diferenciada ainda pode ser estacionária, o que, por sua vez, afetará os parâmetros do modelo.
Então, como determinar a ordem correta de diferenciação?

A ordem correta de diferenciação é a diferenciação mínima necessária para obter uma série quase estacionária que circula em torno de uma média definida e o gráfico ACF chega a zero rapidamente.

Se as autocorrelações forem positivas para muitos atrasos (10 ou mais), a série precisará ser diferenciada. Por outro lado, se a autocorrelação lag 1 em si for muito negativa, a série provavelmente será super diferenciada.

Se não for possível realmente decidir entre duas ordens de diferenciação, então escolha a ordem que apresenta o menor desvio padrão na série diferenciada.

Primeiro, vou verificar se a série é estacionária usando o teste Augmented Dickey Fuller (adfuller()), do pacote statsmodels.

Por quê?

Porque você precisa diferenciar apenas se a série não for estacionária. Senão, nenhuma diferenciação é necessária, ou seja, d = 0.

A hipótese nula do teste ADF é que a série temporal não é estacionária. Portanto, se o valor-p do teste for menor que o nível de significância (0,05), você rejeitará a hipótese nula e poderá inferir que a série temporal é realmente estacionária.

Portanto, se o valor-p > 0,05, prosseguimos em busca da ordem da diferenciação. 

```python
# Teste ADF
resultado = adfuller(dados_treino['Count'])
print('Estatística ADF: %f' % resultado[0])
print('Valor-p: %f' % resultado[1])
```

O valor-p é menor que 0.05, portanto, não precisamos calcular o valor de 'd'. Ou seja, d = 0 no Modelo ARIMA.

Entretanto, vamos fazer mais uma checagem. Vamos aplicar a diferenciação duas vezes a nossa série e comparar os gráficos ACF.

A função np.diff() aplica a diferenciação.

```python
# Área de Plotagem
fig, axes = plt.subplots(3, 2, sharex = True)

# Série Original
axes[0, 0].plot(target.values); axes[0, 0].set_title('Série Original')
plot_acf(target.values, lags = 150, ax = axes[0, 1])

# Diferenciação de Primeira Ordem
axes[1, 0].plot(np.diff(target.values)); axes[1, 0].set_title('Diferenciação de Primeira Ordem')
plot_acf(np.diff(target.values), lags = 150, ax = axes[1, 1])

# Diferenciação de Segunda Ordem
axes[2, 0].plot(np.diff(np.diff(target.values))); axes[2, 0].set_title('Diferenciação de Segunda Ordem')
plot_acf(np.diff(np.diff(target.values)), lags = 150, ax = axes[2, 1])

plt.show()
```

## Determinando o Valor do Parâmetro p para o Modelo ARIMA

A próxima etapa é identificar se o modelo precisa de termos AR. Você pode descobrir o número necessário de termos AR, inspecionando o gráfico PACF (Partial Autocorrelation). Mas o que é PACF?

A autocorrelação parcial pode ser imaginada como a correlação entre a série e seu atraso, após excluir as contribuições dos atrasos intermediários. Portanto, o PACF meio que transmite a correlação pura entre um atraso e a série. Dessa forma, você saberá se esse atraso é necessário no termo AR ou não.

Qualquer autocorrelação em uma série estacionarizada pode ser retificada adicionando termos AR suficientes. Portanto, inicialmente consideramos a ordem do termo AR igual a tantas defasagens que ultrapassam o limite de significância no gráfico PACF.

### Grid Search Método 1 - Stepwise Search 

Usando um Modelo Auto-Arima para retornar os melhores parâmetros de ordem da série, para o menor valor possível da Estatística AIC.

Fazer uma análise manual completa de séries temporais pode ser uma tarefa tediosa, especialmente quando você tem muitos conjuntos de dados para analisar. É preferível automatizar a tarefa de seleção de modelo com a pesquisa em grade (Grid Search). Para o SARIMA, como temos muitos parâmetros, a pesquisa em grade pode levar horas para ser concluída em um conjunto de dados se definirmos o limite de cada parâmetro muito alto. Definir limites muito altos também tornará seu modelo muito complexo e superestimará os dados de treinamento.

Para evitar o longo tempo de execução e o problema de sobreajuste (overfitting), aplicamos o que é conhecido como princípio de parcimônia, onde criamos uma combinação de todos os parâmetros tais que p + d + q + P + D + Q ≤ 6. Outra abordagem é definir cada parâmetro como 0 ou 1 ou 2 e fazer a pesquisa na grade usando o AIC em cada combinação. 

Usaremos a segunda opção, chamada Grid Search Stepwise. Vou definir limites pequenos para os hiperparâmetros, mas você pode testar outros valores se desejar.

```python
# Buscando pela ordem ideal para o modelo
# A função pm.auto_arima aplica o Grid Search e retorna o melhor modelo
modelo_v1 = pm.auto_arima(trainset,
                          seasonal = False, 
                          m = 12,
                          d = 0, 
                          D = 1, 
                          max_p = 2, 
                          max_q = 2,
                          trace = True,
                          error_action = 'ignore',
                          suppress_warnings = True) 
```

Por que a área cinza aumenta? Porque quanto maior o horizonte de previsão, maior a incerteza das previsões. Veja que estamos fazendo previsões de vendas para 5 anos (60 passos no tempo em nossa série), o que aumenta a incerteza a cada novo passo de tempo previsto.

```python
# Teste de Ljung-Box
resultado_teste = sms.diagnostic.acorr_ljungbox(modelo_sarima_v1_fit.resid, lags = [30], boxpierce = False)
print('Valor-p =', resultado_teste)
```

Observe que o valor-p é maior que 0,05, o que indica que os resíduos são independentes no nível de 95%. 

E, portanto, um modelo SARIMA(0, 0, 0)x(0, 1, 1, 12) fornece um bom ajuste e boa capacidade de previsão.
